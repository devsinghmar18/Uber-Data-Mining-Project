```python
df = pd.read_csv("uber_data.csv")
```

```python
df.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'VendorID', 'pickup_longitude', 'pickup_latitude', 'RatecodeID', 'store_and_fwd_flag', 'dropoff_longitude', 'dropoff_latitude', 'payment_type'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.head(15)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
X = df[['passenger_count', 'trip_distance', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge']]
y_fare = df['total_amount']  # Fare amount is the total amount in this case
y_tip = df['tip_amount']
```

```python
# Split the dataset into train and test sets
X_train, X_test, y_fare_train, y_fare_test, y_tip_train, y_tip_test = train_test_split(X, y_fare, y_tip, test_size=0.2, random_state=42)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the SVR model for fare amount prediction
fare_svr = SVR(kernel='linear')
fare_svr.fit(X_train_scaled, y_fare_train)
```

```python
# Predict fare amount
fare_predictions = fare_svr.predict(X_test_scaled)
```

```python
fare_predictions
```

```python
# Train the SVR model for tip amount prediction
tip_svr = SVR(kernel='linear')
tip_svr.fit(X_train_scaled, y_tip_train)
```

```python
# Predict tip amount
tip_predictions = tip_svr.predict(X_test_scaled)
```

```python
tip_predictions
```

```python
# Evaluate the models
fare_rmse = mean_squared_error(y_fare_test, fare_predictions, squared=False)
tip_rmse = mean_squared_error(y_tip_test, tip_predictions, squared=False)

print("Fare amount RMSE:", fare_rmse)
print("Tip amount RMSE:", tip_rmse)
```

```python
import matplotlib.pyplot as plt

# Plot for fare amount prediction
plt.figure(figsize=(10, 5))
plt.scatter(y_fare_test, fare_predictions, color='blue')
plt.plot([y_fare_test.min(), y_fare_test.max()], [y_fare_test.min(), y_fare_test.max()], 'k--', lw=2)
plt.xlabel('Actual Fare Amount')
plt.ylabel('Predicted Fare Amount')
plt.title('Actual vs Predicted Fare Amount')
plt.grid(True)
plt.show()

# Plot for tip amount prediction
plt.figure(figsize=(10, 5))
plt.scatter(y_tip_test, tip_predictions, color='red')
plt.plot([y_tip_test.min(), y_tip_test.max()], [y_tip_test.min(), y_tip_test.max()], 'k--', lw=2)
plt.xlabel('Actual Tip Amount')
plt.ylabel('Predicted Tip Amount')
plt.title('Actual vs Predicted Tip Amount')
plt.grid(True)
plt.show()
```
